```python
import pandas as pd
```

### Đọc dữ liệu và khảo sát

```python
#1, Đọc dữ liệu và khảo sát ban đầu
customer = pd.read_csv("customer_info.csv")
invoice_header = pd.read_csv("invoice_header.csv")
invoice_detail = pd.read_csv("invoice_detail.csv")

print(customer.info())
print(invoice_header.info())
print(invoice_detail.info())
```

### Chuẩn hóa, xử lý các giá trị thiếu và bất thường

```python
#làm sạch customer_info.csv
#chuẩn hóa customer_id
customer["customer_id"] = customer["customer_id"].astype(str).str.strip().str.upper()
#chuẩn hóa họ tên khách hàng
customer["full_name"] = (
    customer["full_name"]
    .astype(str)
    .str.strip()
    .str.replace(r"\s+", " ")
    .str.title()
)
#chuẩn hóa khu vực
customer["region"] = customer["region"].astype(str).str.strip().str.title()

customer["region"] = customer["region"].replace({
    "Bac": "Bắc",
    "Trung": "Trung",
    "Nam": "Nam"
})

customer["region"] = customer["region"].replace("Nan", pd.NA)
customer["region"] = customer["region"].fillna("Không xác định")
#chuẩn hóa ngày sinh
customer["dob"] = pd.to_datetime(
    customer["dob"],
    format="%d/%m/%Y",
    errors="coerce"
)
```

```python
# làm sạch invoice_header.csv
#chuẩn hóa mã khách hàng
invoice_header["customer_id"] = (invoice_header['customer_id'].astype(str).str.strip().str.upper())
#chuẩn hóa ngày mua hàng
invoice_header["date"] = pd.to_datetime(invoice_header["date"], errors = "coerce", format="%d/%m/%Y")
```

```python
#làm sạch invoice_detail.csv
#làm sạch tên sản phẩm
invoice_detail["product_name"] = (invoice_detail["product_name"].astype(str).str.strip().str.replace(r"\s+", " ").str.title())
#làm sạch số lượng
invoice_detail["quantity"] = pd.to_numeric(invoice_detail["quantity"], errors = "coerce")
invoice_detail.loc[invoice_detail["quantity"] <= 0, "quantity"] = pd.NA
#làm sạch đơn giá
invoice_detail["unit_price"] = pd.to_numeric(invoice_detail["unit_price"], errors = "coerce")
invoice_detail.loc[invoice_detail["unit_price"] <= 0, "unit_price"] = pd.NA
```

```python
print(customer.isnull().sum())
print(invoice_header.isnull().sum())
print(invoice_detail.isnull().sum())
```

### Xuất dữ liệu sạch

```python
customer.to_csv("customer_info_clean.csv", index=False)
invoice_header.to_csv("invoice_header_clean.csv", index=False)
invoice_detail.to_csv("invoice_detail_clean.csv", index=False)
```
