```python
import pandas as pd

customer = pd.read_csv("../data_raw/customer_info.csv")
invoice_h = pd.read_csv("../data_raw/invoice_header.csv")
invoice_d = pd.read_csv("../data_raw/invoice_detail.csv")

customer['customer_id'] = customer['customer_id'].astype(str).str.strip()
customer['full_name'] = customer['full_name'].astype(str).str.strip().str.title()
customer['region'] = customer['region'].astype(str).str.strip().str.capitalize()
customer['region'] = customer['region'].replace({'Bac':'Bắc','Trung':'Trung','Nam':'Nam'})
customer['region'] = customer['region'].fillna('Unknown')
customer['dob'] = pd.to_datetime(customer['dob'], errors='coerce')

invoice_h['invoice_id'] = invoice_h['invoice_id'].astype(str).str.strip()
invoice_h['customer_id'] = invoice_h['customer_id'].astype(str).str.strip()
invoice_h['date'] = pd.to_datetime(invoice_h['date'], errors='coerce')

invoice_d['product_name'] = invoice_d['product_name'].astype(str).str.strip().str.title()
invoice_d['quantity'] = pd.to_numeric(invoice_d['quantity'], errors='coerce')
invoice_d = invoice_d[invoice_d['quantity'] > 0]

invoice_d['unit_price'] = (
    invoice_d['unit_price']
    .astype(str)
    .str.replace('[^0-9.]', '', regex=True)
)
invoice_d['unit_price'] = pd.to_numeric(invoice_d['unit_price'], errors='coerce')
invoice_d = invoice_d[invoice_d['unit_price'] > 0]

customer.to_csv("../data_clean/customer_info_clean.csv", index=False)
invoice_h.to_csv("../data_clean/invoice_header_clean.csv", index=False)
invoice_d.to_csv("../data_clean/invoice_detail_clean.csv", index=False)

print("DONE: data_clean đã được tạo")
```
