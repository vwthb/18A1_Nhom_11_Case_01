Đọc file

```python
import pandas as pd

khach_hang = pd.read_csv("customer_info.csv")
hoa_don = pd.read_csv("invoice_header.csv")
chi_tiet = pd.read_csv("invoice_detail.csv")
```

 Lọc khách hàng theo khu vực (Bắc / Trung / Nam).

```python
# Khách hàng khu vực Bắc
khach_hang_bac = khach_hang[khach_hang["region"] == "Bắc"]

# Khách hàng khu vực Trung
khach_hang_trung = khach_hang[khach_hang["region"] == "Trung"]

# Khách hàng khu vực Nam
khach_hang_nam = khach_hang[khach_hang["region"] == "Nam"]

print(khach_hang_bac)
print(khach_hang_trung)
print(khach_hang_nam)
```

Thống kê số lượng đơn hàng theo từng khách hàng.

```python
so_don_theo_khach = (
    hoa_don
    .groupby("customer_id")["invoice_id"]
    .count()
    .reset_index(name="so_luong_don_hang")
)

print(so_don_theo_khach)
```

Thống kê số khách hàng theo khu vực

```python
so_khach_theo_khu_vuc = (
    khach_hang
    .groupby("region")["customer_id"]
    .count()
    .reset_index(name="so_luong_khach_hang")
)

print(so_khach_theo_khu_vuc)
```

Tìm các đơn hàng không có chi tiết sản phẩm

```python
don_khong_chi_tiet = hoa_don[
    ~hoa_don["invoice_id"].isin(chi_tiet["invoice_id"])
]

print(don_khong_chi_tiet)
```

Thống kê sản phẩm được mua nhiều nhất.

```python
san_pham_ban_chay = (
    chi_tiet
    .groupby("product_name")["quantity"]
    .sum()
    .reset_index(name="tong_so_luong")
    .sort_values(by="tong_so_luong", ascending=False)
)

san_pham_mua_nhieu_nhat = san_pham_ban_chay.head(1)

print(san_pham_mua_nhieu_nhat)
```
